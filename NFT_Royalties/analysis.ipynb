## Royalties Effect on NFTs
#### This is an analysis of 142 Major projects and the effect that royalties have on the estimated market cap of these projects

```python
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.read_csv('data.csv')
df = pd.DataFrame(data)
max_royalty = df['royalties'].max()
print(max_royalty)
print(df.shape)
print(df.columns)
df_g = df[df.assets != df.assets.max()]
```

```python
sns.lmplot(x='royalties',y='mkt_cap',data=df_g,fit_reg=True) 
sns.lmplot(x='royalties',y='volume',data=df_g,fit_reg=True) 
sns.lmplot(x='royalties',y='floor',data=df_g,fit_reg=True) 
sns.lmplot(x='royalties',y='sales',data=df_g,fit_reg=True) 
sns.lmplot(x='royalties',y='avg_price',data=df_g,fit_reg=True) 
sns.lmplot(x='royalties',y='assets',data=df_g,fit_reg=True) 
```

```python
grid = sns.lmplot(x='royalties', y='mkt_cap', data=df_g, fit_reg=True) 
grid.set(yscale="log")

grid = sns.lmplot(x='royalties', y='volume', data=df_g, fit_reg=True) 
grid.set(yscale="log")

grid = sns.lmplot(x='royalties', y='floor', data=df_g, fit_reg=True) 
grid.set(yscale="log")
```

## Regression Models and Coerrelation
#### Market Cap
Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt



from scipy.stats import pearsonr, spearmanr

X=df['royalties'].values.reshape(-1,1) # this  has shape (XXX, 1) - it's 2D
Y=df['mkt_cap'].values.reshape(-1,1) # this  has shape (XXX, 1) - it's 2D
model = LinearRegression().fit(X,Y)

r_sq = model.score(X, Y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")

```

Testing the Regression

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)

linreg=LinearRegression()
linreg.fit(x_train,y_train)
y_pred=linreg.predict(x_test)
Accuracy=r2_score(y_test,y_pred)*100
print(" Accuracy of the model is %.2f" %Accuracy)
```

```python
plt.scatter(y_test,y_pred);
plt.xlabel('Actual');
plt.ylabel('Predicted');
sns.regplot(x=y_test,y=y_pred,ci=None,color ='red');
```
